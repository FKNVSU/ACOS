Задание: визуализация частичных сумм Фурье.

1)для ступенчатой(f = 1)

2)для треуольника(f = pi - |x|, -pi < x < pi else 0)

3)для линейной(f = x, -5 < x < 5)

n - количество слагаемых
и чтобы можно было задавать интервал


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
class Visualization:
    def __init__(self, a, b, n):
        self.a = a
        self.b = b
        self.n = n
        self.x = np.linspace(self.a, self.b, n+1)
        self.y = []

    def plot(self, show_linear_func=False):
        fig, ax = plt.subplots(figsize=(12, 6))
        ax.plot(self.x, self.y)
        if show_linear_func: ax.plot(np.linspace(self.a, self.b, self.n), np.linspace(self.a, self.b, self.n))
        plt.show()

    def s_step_function(self):
        self.y = 0
        for k in range(1, self.n + 1):
            self.y += 2 * np.sin(k) / (np.pi * k) * np.cos(k * self.x)
        self.y += 1 / np.pi
        return self.y

    def s_triangular_impulse(self):
        self.y = 0
        for k in range(1, self.n + 1):
            self.y += (2 / (np.pi * k**2) * ( 1 - (-1)**k )) * np.cos(k * self.x)
        self.y += np.pi / 2
        return self.y

    def s_linear_function(self):
        self.y = 0
        for k in range(1, self.n + 1):
            self.y += (-10 * np.cos(np.pi * k) * np.sin(np.pi * k * self.x / 5)) / (np.pi * k)
```

```python
v1 = Visualization(-2, 2, 200)
v1.s_step_function()
v1.plot()
```

```python
v2 = Visualization(-np.pi, np.pi, 200)
v2.s_triangular_impulse()
v2.plot()
```

```python
v3 = Visualization(-7, 7, 100)
v3.s_linear_function()
v3.plot(show_linear_func=True)
```
