```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot(x, y, a=None, b=None, n=None):
    fig, ax = plt.subplots(figsize=(12, 6))
    ax.plot(x, y)
    if a is not None:
        linear_func = np.linspace(a, b, n)
        ax.plot(linear_func, linear_func)
    plt.show()

def s_step_function(a, b, n):
    x = np.linspace(a, b, n+1)
    y = 0
    for k in range(1, n + 1):
        y += 2 * np.sin(k) / (np.pi * k) * np.cos(k * x)
    y += 1 / np.pi
    plot(x, y)

def s_triangular_impulse(a, b, n):
    x = np.linspace(a, b, n+1)
    y = 0
    for k in range(1, n + 1):
        y += (2 / (np.pi * k**2) * ( 1 - (-1)**k )) * np.cos(k * x)
    y += np.pi / 2
    plot(x, y)

def s_linear_function(a, b, n):
    x = np.linspace(a, b, n+1)
    y = 0
    for k in range(1, n + 1):
        y += (-10 * np.cos(np.pi * k) * np.sin(np.pi * k * x / 5)) / (np.pi * k)
    plot(x, y, a, b, n)
```

```python
s_step_function(-2, 2, 100)
```

```python
s_triangular_impulse(-np.pi, np.pi, 200)
```

```python
s_linear_function(-7, 7, 100)
```
